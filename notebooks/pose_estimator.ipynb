```python
import os
import cv2
import numpy as np
from tqdm import tqdm
import romp

settings = romp.main.default_settings
settings.mode = 'image'
settings.calc_smpl = True
romp_model = romp.ROMP(settings)


input_dir = '/workspace/HRHumanTex/scripts/SHHQ_testdata/images'
output_dir = '/workspace/HRHumanTex/scripts/SHHQ_testdata/pose_output_npz'
os.makedirs(output_dir, exist_ok=True)


for fname in tqdm(os.listdir(input_dir)):
    if not fname.endswith('.jpg'):
        continue
    image_path = os.path.join(input_dir, fname)
    image = cv2.imread(image_path)

    if image is None:
        print(f"Failed to load {fname}")
        continue

    results = romp_model(image)
    
 
    if 'id_' in fname:
        image_id = fname.split('id_')[-1].split('-')[0]
    else:
        image_id = os.path.splitext(fname)[0]
        
    out_path = os.path.join(output_dir, f'{image_id}.npz')
    np.savez_compressed(out_path, results=results)

print("✅ All images processed and saved as .npz.")
```
