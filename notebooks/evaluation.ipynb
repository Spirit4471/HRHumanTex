```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv("metrics1024inpainting.csv", index_col="model")
models = df.index.tolist()


fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Different model evaluated on 4 metrics", fontsize=16)


metrics = [
    ("psnr", "PSNR (↑)"),
    ("ssim", "SSIM (↑)"),
    ("lpips", "LPIPS (↓)"),
    ("fid",       "FID (↓)")
]
x = np.arange(len(models))
width = 0.8

for ax, (col, title) in zip(axes.flatten(), metrics):
    vals = df[col].values
    bars = ax.bar(x, vals, width, color="C0", edgecolor="k")
    ax.set_xticks(x)
    ax.set_xticklabels(models, rotation=45, ha="right")
    ax.set_title(title)
    ax.grid(axis="y", linestyle="--", alpha=0.5)

    for bar, v in zip(bars, vals):
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            v + (0.01 * (vals.max() - vals.min())),
            f"{v:.3f}",
            ha="center",
            va="bottom",
            fontsize=8
        )

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig("comparison_4metrics.png", dpi=200)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv("metrics1024inpainting.csv").set_index("model")
metrics = ["psnr","ssim","lpips","fid"]
inverse_metrics = ['lpips', 'fid']


ranges = df[metrics].max() - df[metrics].min()
ranges[ranges == 0] = 1 
df_norm = (df[metrics] - df[metrics].min()) / ranges
df_norm[inverse_metrics] = 1 - df_norm[inverse_metrics]  


plt.figure(figsize=(10, 10))
N = len(metrics)
angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
angles += angles[:1]

ax = plt.subplot(polar=True)
for model in df_norm.index:
    vals = df_norm.loc[model].tolist()
    vals += vals[:1]
    ax.plot(angles, vals, linewidth=2, label=model)
    ax.fill(angles, vals, alpha=0.1)

ax.set_xticks(angles[:-1])
ax.set_xticklabels([m.upper() for m in metrics], fontsize=12)
ax.yaxis.grid(color='gray', linestyle='--')
ax.set_rlabel_position(45)
ax.set_title("Model Performance Comparison", pad=20, fontsize=14)
plt.legend(loc='upper left', bbox_to_anchor=(1.1, 1.1))
plt.tight_layout()
plt.savefig("radar_comparison.png", dpi=300, bbox_inches='tight')
```

```python
import numpy as np
import torch
from smplx import SMPL
from pytorch3d.structures import Meshes
from pytorch3d.renderer import (
    OrthographicCameras, RasterizationSettings,
    MeshRenderer, MeshRasterizer, SoftPhongShader,
    PointLights, BlendParams, look_at_view_transform
)
from pytorch3d.renderer.mesh.textures import TexturesVertex
import matplotlib.pyplot as plt

# ------------------------
# Step 1: Load .npz
# ------------------------
npz_path = "/workspace/HRHumanTex/scripts/test_data/pose_output_npz/00002565.npz"  # e.g., id_00012345_smpl_params.npz
data = np.load(npz_path, allow_pickle=True)
if 'results' in data.files:
    smpl_data = data['results'].item()
else:
    smpl_data = dict(data)

thetas_np = np.array(smpl_data['smpl_thetas']).reshape(-1)
betas_np = np.array(smpl_data.get('smpl_betas', np.zeros(10,))).reshape(1, 10)

# ------------------------
# Step 2: Prepare SMPL inputs
# ------------------------
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
smpl_model = SMPL(model_path="/workspace/HRHumanTex/sample-data/SMPL/models/smpl", gender="neutral").to(device)

thetas = torch.tensor(thetas_np, dtype=torch.float32).unsqueeze(0).to(device)
global_orient = thetas[:, :3]
body_pose = thetas[:, 3:]
betas = torch.tensor(betas_np, dtype=torch.float32).to(device)

# ------------------------
# Step 3: Forward SMPL to get mesh
# ------------------------
output = smpl_model(betas=betas, body_pose=body_pose, global_orient=global_orient, return_verts=True)
verts = output.vertices[0]
faces = torch.tensor(smpl_model.faces.astype(np.int64)).to(device)

# ------------------------
# Step 4: Build white mesh
# ------------------------
verts_rgb = torch.ones_like(verts)[None]  # white color
textures = TexturesVertex(verts_features=verts_rgb.to(device))
mesh = Meshes(verts=[verts], faces=[faces], textures=textures)

# ------------------------
# Step 5: Renderer setup
# ------------------------
R, T = look_at_view_transform(dist=2.0, elev=0.0, azim=180)
cameras = OrthographicCameras(device=device, R=R, T=T)
lights = PointLights(
    device=device,
    location=[[0.0, 0.5, -2.0]],
    ambient_color=[[0.6, 0.6, 0.6]],
    diffuse_color=[[0.4, 0.4, 0.4]],
    specular_color=[[0.1, 0.1, 0.1]],
)

raster_settings = RasterizationSettings(image_size=512)
blend_params = BlendParams(background_color=(0.0, 0.0, 0.0))

renderer = MeshRenderer(
    rasterizer=MeshRasterizer(cameras=cameras, raster_settings=raster_settings),
    shader=SoftPhongShader(device=device, cameras=cameras, lights=lights, blend_params=blend_params)
)

# ------------------------
# Step 6: Render and Save
# ------------------------
img = renderer(mesh)[0, ..., :3].cpu().detach().numpy()
img = np.flipud(img)
img = np.fliplr(img)
plt.imshow(img)
plt.axis("off")
plt.title("SMPL White Mesh")
plt.savefig("white_smpl.png", dpi=300, bbox_inches='tight')
plt.show()
```
